```python
import numpy as np
import pandas as pd
import os
```

```python
experiment_dir = 'data/equalPrecisions_N20_n1000'
```

```python
data = []

for file in os.listdir(os.path.join(experiment_dir, 'data')):
    fname = os.path.join(experiment_dir, 'data', file)

    data.append(pd.read_csv(fname))

agg_df = pd.concat(data, ignore_index=True)
agg_df.to_csv(os.path.join(experiment_dir, f'{os.path.basename(experiment_dir)}.csv'), index=False)
agg_df
```
